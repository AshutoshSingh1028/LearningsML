```python
from sklearn.datasets import load_diabetes
```

```python
dataset=load_diabetes()
```

```python
dataset
```

```python
print(dataset['DESCR'])
```

```python
import pandas as pd
df=pd.DataFrame(dataset.data,columns=['age',
  'sex',
  'bmi',
  'bp',
  's1',
  's2',
  's3',
  's4',
  's5',
  's6'])
```

```python
# independent and dependent features
X=df
y=dataset['target']
```

```python
df.head()
```

```python
# splitting
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
X_train.head()
```

```python
X_train.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(15,11))
sns.heatmap(X_train.corr(),annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor()
```

```python
print(X_train.shape)  # Should print (n_samples, n_features)
print(y_train.shape)  # Should print (n_samples,)
```

```python
regressor.fit(X_train,y_train)
```

## Hyperparameter Tuning

```python
params={
    'criterion':['squared_error', 'friedman_mse', 'absolute_error'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5,10,15,20,25],
    'max_features':['auto','sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
regressor=DecisionTreeRegressor()
```

```python
grid=GridSearchCV(regressor,param_grid=params,cv=5,scoring='neg_mean_squared_error')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
model=DecisionTreeRegressor(criterion='friedman_mse',max_depth=4,max_features='sqrt',splitter='random')
```

```python
model.fit(X_train,y_train)
```

```python
# Visualize the Decision Tree
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(model,filled=True)
```
